```python
import pandas as pd
import matplotlib.pyplot as plt

# Ensure datetime conversion
df['Created Date'] = pd.to_datetime(df['Created Date'])
df['Closed Date'] = pd.to_datetime(df['Closed Date'])

# Calculate closing time in hours
df['Closing_Hours'] = (df['Closed Date'] - df['Created Date']).dt.total_seconds() / 3600

# Extract day of the week
df['Weekday'] = df['Created Date'].dt.day_name()

# Calculate average closing hours per weekday in proper order
weekday_order = ['Sunday', 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday']
average_closing = df.groupby('Weekday')['Closing_Hours'].mean().reindex(weekday_order)

# Plotting
average_closing.plot(kind='bar', color='red')
plt.title("Average Closing Time of Requests by Weekday", fontsize=15)
plt.ylabel("Avg Time to Close (Hours)", fontsize=13)
plt.xlabel("Day of the Week", fontsize=13)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show() 
```

```python
# Create a new column that captures month and year from the Created Date
df['Month_Year_Label'] = df['Created Date'].dt.to_period('M')

# Count the number of requests per month-year
monthly_counts = df.groupby('Month_Year_Label').size()

# Set up the plot dimensions and create the line chart
fig, axis = plt.subplots(figsize=(14, 5))  
axis.plot(monthly_counts.index.astype(str), monthly_counts.values, color='steelblue')

# Add chart title and axis labels with updated font sizes
axis.set_title("311 Request Volume by Month", fontsize=18)
axis.set_xlabel("Month and Year", fontsize=13)
axis.set_ylabel("Request Count", fontsize=13)

# Improve x-axis label readability and adjust spacing
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Convert created and closed dates to datetime if not already
df['Created Date'] = pd.to_datetime(df['Created Date'])
df['Closed Date'] = pd.to_datetime(df['Closed Date'])

# Calculate time taken to close the request in hours
df['Closing_Hours'] = (df['Closed Date'] - df['Created Date']).dt.total_seconds() / 3600

# Group by both complaint type and borough to compute average closing time
average_closing = df.groupby(['Complaint Type', 'Borough'])['Closing_Hours'].mean().reset_index()

# Reshape the data to prepare it for horizontal bar plot
pivoted_data = average_closing.pivot(index='Complaint Type', columns='Borough', values='Closing_Hours')

# Create the plot
pivoted_data.plot(kind='barh', figsize=(14, 9), colormap='coolwarm')

# Add a descriptive title and axis labels with varied font sizes
plt.title("Avg. Request Closure Time by Complaint and Borough", fontsize=18, color='navy')
plt.xlabel("Average Time (hrs)", fontsize=14, color='darkgreen')
plt.ylabel("Complaint Category", fontsize=14, color='maroon')

# Customize the legend and layout
plt.legend(title='Borough', bbox_to_anchor=(1.02, 1), loc='upper left', fontsize=10, title_fontsize=11)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.tight_layout()
plt.show()
```

```python
# Ensure 'Created Date' and 'Closed Date' are in datetime format
df['Created Date'] = pd.to_datetime(df['Created Date'])
df['Closed Date'] = pd.to_datetime(df['Closed Date'])

# Calculate the time difference between 'Closed Date' and 'Created Date' in days
df['Request_Closing_Time_Days'] = (df['Closed Date'] - df['Created Date']).dt.days

# Select relevant columns and remove rows with missing values
avg_response = df[['Complaint Type', 'Request_Closing_Time_Days']].dropna()

# Group the data by 'Complaint Type'
grouped = avg_response.groupby('Complaint Type')

# Create a list of groups where each group contains the 'Request_Closing_Time_Days' values for each complaint type
groups = [group['Request_Closing_Time_Days'].values for name, group in grouped if len(group) > 1]

# Perform the ANOVA test to check if there's a significant difference between the means of the groups
from scipy.stats import f_oneway
f_stat, p_val = f_oneway(*groups)

# Print the F-statistic and p-value, rounded to an appropriate number of decimal places
print("F-statistic:", round(f_stat, 4))
print("p-value:", round(p_val, 4))

# Check if the p-value is less than the significance level (0.05), and make a decision about the null hypothesis
if p_val < 0.05:
    print("Reject the null hypothesis: The average response time differs across complaint types.")
else:
    print("Accept the null hypothesis: No significant difference in response time among complaint types.")
```

```python
# Import the chi2_contingency function from scipy.stats
from scipy.stats import chi2_contingency

# Create a contingency table to observe the relationship between 'Complaint Type' and 'Location Type'
contingency_table = pd.crosstab(df['Complaint Type'], df['Location Type'])

# Perform Chi-Square test on the contingency table to check for independence
test_statistic, p_value, degrees_of_freedom, expected_values = chi2_contingency(contingency_table)

# Output the results of the Chi-Square test
print("Chi-Square Statistic:", test_statistic)
print("P-value:", p_value)

# Interpret the p-value and decide whether to reject the null hypothesis
if p_value <= 0.05:
    print("Reject the null hypothesis")
else:
    print("Accept the null hypothesis")
```
